```python
df=pd.read_csv("supermarkets/supermarkets.csv")
```

```python
df
```

```python
df.iloc[0]['City']
```

```python
def geo(row):
    s=row['Address']+','+row['City']+","+row['State']
    print("s",s)
    loc=nom.geocode(s)
    return (loc.latitude,loc.longitude)
```

```python
df['Coordinate']=df.apply(geo,axis=1)
```

```python
df['latitude']=df["Coordinate"].apply(lambda x:x[0])
```

```python
df['longtitude']=df['Coordinate'].apply(lambda x:x.longtitude if x!=None else None)
```
